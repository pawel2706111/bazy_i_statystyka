<h3>Zadanie 1</h3>

```python
import numpy as np

wartosc = np.array(list(range(1, 7)))
prawdopodobienstwo = np.array([1/6 for i in range(0, 6)])

srednia = np.sum(wartosc * prawdopodobienstwo)
wariancja = np.sum(prawdopodobienstwo * (wartosc - srednia)**2)
odchylenie_standardowe = np.sqrt(wariancja)

print('Średnia:', srednia)
print('Wariancja:', wariancja)
print('Odchylenie standardowe:', odchylenie_standardowe)
```

<h3>Zadanie 2</h3>

```python
import scipy.stats as scs

rozklad_bernoulliego = scs.bernoulli.rvs(0.3, size=100)

rozklad_dwumianowy = scs.binom.rvs(10, 0.3, size=100)

rozklad_poissona = scs.poisson.rvs(3, size=100)

print('Rozkład Bernoulliego:')
print(rozklad_bernoulliego)

print('\nRozkład Dwumianowy:')
print(rozklad_dwumianowy)

print('\nRozkład Poissona:')
print(rozklad_poissona)
```

<h3>Zadanie 3</h3>

```python
srednia, wariancja, skosnosc, kurtoza = scs.bernoulli.stats(0.3, moments='mvsk')

print('Rozkład Bernoulliego')
print('Średnia:', srednia)
print('Wariancja:', wariancja)
print('Skosność:', skosnosc)
print('Kurtoza:', kurtoza)

srednia, wariancja, skosnosc, kurtoza = scs.binom.stats(10, 0.3, moments='mvsk')

print('\nRozkład Dwumianowy')
print('Średnia:', srednia)
print('Wariancja:', wariancja)
print('Skosność:', skosnosc)
print('Kurtoza:', kurtoza)

srednia, wariancja, skosnosc, kurtoza = scs.poisson.stats(3, moments='mvsk')

print('\nRozkład Poissona')
print('Średnia:', srednia)
print('Wariancja:', wariancja)
print('Skosność:', skosnosc)
print('Kurtoza:', kurtoza)
```

<h3>Zadanie 4</h3>

```python
import matplotlib.pyplot as plt

x = np.array([0, 1])
bernoulli_pmf = scs.bernoulli.pmf(x, 0.3)
plt.bar(x, bernoulli_pmf, edgecolor='black')
plt.title('Rozkład Bernoulliego')
plt.xlabel('Wartość')
plt.ylabel('Prawdopodobieństwo')
plt.show()
```

```python
x = np.arange(0, 11)
binomial_pmf = scs.binom.pmf(x, 10, 0.3)
plt.bar(x, binomial_pmf, edgecolor='black')
plt.title('Rozkład Dwumianowy')
plt.xlabel('Wartość')
plt.ylabel('Prawdopodobieństwo')
plt.show()
```

```python
x = np.arange(0, 15)
poisson_pmf = scs.poisson.pmf(x, 3)
plt.bar(x, poisson_pmf, edgecolor='black')
plt.title('Rozkład Poissona')
plt.xlabel('Wartość')
plt.ylabel('Prawdopodobieństwo')
plt.show()
```

<h3>Zadanie 6</h3>

```python
import pandas as pd

srednia = 0
odchylenie = 2
n = 100

dane = np.random.normal(srednia, odchylenie, n)

srednia_rel = np.mean(dane)
wariancja_rel = np.var(dane)
odchylenie_rel = np.std(dane)
skosnosc_rel = scs.skew(dane)
kurtoza_rel = scs.kurtosis(dane)

df_indeksy = ['Średnia', 'Wariancja', 'Odchylenie standardowe', 'Skośność', 'Kurtoza']
df_dane = {'Statystyka Rzeczywista': [srednia_rel, wariancja_rel, odchylenie_rel, skosnosc_rel, kurtoza_rel],
           'Statystyka Teoretyczna': [srednia, odchylenie, odchylenie**2, 0, 0]}
df = pd.DataFrame(data=df_dane, index=df_indeksy)
print(f'Statystyki dla n = {n}:')
display(df)

n = 10000

dane = np.random.normal(srednia, odchylenie, n)

srednia_rel = np.mean(dane)
wariancja_rel = np.var(dane)
odchylenie_rel = np.std(dane)
skosnosc_rel = scs.skew(dane)
kurtoza_rel = scs.kurtosis(dane)

statystyka_rel = [srednia_rel, wariancja_rel, odchylenie_rel, skosnosc_rel, kurtoza_rel]
df['Statystyka Rzeczywista'] = statystyka_rel
print(f'\nStatystyki dla n = {n}:')
display(df)
```
